```python


# Sort the DataFrame by "Score" in descending order to get the top 10 countries
top_10_countries = happiness_report.sort_values(by='Score', ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(10, 8))
plt.bar(top_10_countries['Country or region'], top_10_countries['Score'], color='skyblue')

# Add labels and title
plt.xlabel('Country or region')
plt.ylabel('Happiness Score')
plt.title('Happiness Score for Top 10 Countries')




# Display the plot
plt.tight_layout()
plt.show()
```
